```python
ev_adoption_year = df["Model Year"].value_counts().sort_index()
```

```python
sns.barplot(x = ev_adoption_year.index , y = ev_adoption_year.values , palette = "viridis")
plt.title("EV Adoption Over Time")
plt.xlabel("Year")
plt.ylabel("Number of vehicles registered")
plt.xticks(rotation = 45)
plt.tight_layout()
```

The bar chart shows a significant increase in EV adoption, particularly from 2017 onwards. The number of registered EVs rose modestly until 2016, then accelerated rapidly, with 2023 marking the highest peak in registrations.

```python
ev_county_dist = df['County'].value_counts()
top_counties = ev_county_dist.head(3).index
top_counties
```

```python
top_counties_data = df[df['County'].isin(top_counties)]
top_counties_data
```

```python
ev_city = top_counties_data.groupby(['County' , 'City']).size().sort_values(ascending=False).reset_index(name='Number of Vehicles')
ev_city
```

```python
top_cities = ev_city.head(10)
sns.barplot(x = 'Number of Vehicles', y = 'City' , hue = 'County' ,data = top_cities, palette = 'rainbow')
plt.title('Top Cities in Top Counties by EV Registrations')
plt.xlabel('Number of Vehicles Registered')
plt.ylabel('City')
plt.legend(title='County')
plt.tight_layout()
```

The chart displays the number of EV registrations in top cities across three counties. Seattle leads significantly with the highest registrations, followed by Bellevue and Redmond. The majority of these cities are in King County, with Bothell in Snohomish County and Tacoma in Pierce County also notable. King County dominates in EV registrations compared to Snohomish and Pierce.

```python
ev_type_dist = df['Electric Vehicle Type'].value_counts()
plt.figure(figsize=(4, 4))
plt.pie(ev_type_dist,labels=ev_type_dist.index, autopct='%1.1f%%', startangle=140)
plt.title('Disrtibution of Electric Vehicle Types')
plt.tight_layout()
```

The above graph shows that BEVs are more popular or preferred over PHEVs among the electric vehicles registered in the United States.

```python
ev_make_dist = df['Make'].value_counts().head(10)

sns.barplot(x= ev_make_dist.values , y = ev_make_dist.index , palette = 'cubehelix')
plt.title('Top 10 Popular EV Makes')
plt.xlabel('Number of Vehicles')
plt.ylabel('Make')
plt.tight_layout()
```

The chart indicates that TESLA has the highest number of registered vehicles, followed by NISSAN and CHEVROLET. FORD, BMW, KIA, TOYOTA, VOLKSWAGEN, JEEP, and HYUNDAI follow in descending order of registrations.

```python
top_3_makes = ev_make_dist.head(3).index

top_makes_data = df[df['Make'].isin(top_3_makes)]
ev_model = top_makes_data.groupby(['Make', 'Model']).size().sort_values(ascending=False).reset_index(name='Number of Vehicles')

top_models = ev_model.head(10)
sns.barplot(x='Number of Vehicles', y='Model', hue='Make', data=top_models, palette="viridis")
plt.title('Top Models in Top 3 Makes by EV Registrations')
plt.xlabel('Number of Vehicles Registered')
plt.ylabel('Model')
plt.legend(title='Make', loc='center right')
plt.tight_layout()
```

The graph shows that TESLA's MODEL Y and MODEL 3 have the highest registrations, with MODEL Y leading. NISSAN's LEAF is the most registered non-TESLA model. TESLA's MODEL S and MODEL X also have significant numbers. CHEVROLET's BOLT EV, VOLT, and BOLT EUV follow, with NISSAN's ARIYA and CHEVROLET's SPARK having the fewest registrations.

```python
sns.histplot(df['Electric Range'], bins = 30, kde = True , color = 'Royalblue')
plt.title('Distribution of Electric Vehicle Ranges')
plt.xlabel('Electric Range (miles)')
plt.ylabel('Number of Vehicles')
plt.axvline(df['Electric Range'].mean(), color='red', linestyle='--', label=f'Mean Range: {df["Electric Range"].mean():.2f} miles')
plt.legend()
```

The graph displays the distribution of the mean electric range among electric vehicles. Key observations include:

- A high frequency of vehicles with a low electric range, with a noticeable peak just before 50 miles.
- The distribution is right-skewed, with a long tail extending towards higher ranges. However, vehicles with higher ranges are much less frequent.
- The mean electric range for this set of vehicles is approximately 58.84 miles, indicating that most vehicles have a relatively low range compared to     the highest values observed in the graph.
- Despite some electric vehicles having ranges up to around 350 miles, the majority are below the mean range.


```python
average_range_by_year = df.groupby('Model Year')['Electric Range'].mean().reset_index()

sns.lineplot(x = 'Model Year', y = 'Electric Range' , data = average_range_by_year , marker = 'o' , color = 'green')
plt.title('Average Electric Range by Model Year')
plt.xlabel('Model Year')
plt.ylabel('Average Electric Range (miles)')
```

The graph illustrates the progression of the average electric range of EVs from around 2000 to 2024. Key findings include:

- A general upward trend in the average range, reflecting technological improvements.
- A peak around 2020, marking the highest average range observed.
- A significant drop after 2020, possibly due to incomplete data or the introduction of lower-range models.
- A slight recovery in the average range in the most recent year.
- Overall, the trend over the last two decades shows an increase in the electric range of EVs.


```python
average_range_by_model = top_makes_data.groupby(['Make', 'Model'])['Electric Range'].mean().sort_values(ascending=False).reset_index()

top_range_models = average_range_by_model.head(10)

barplot = sns.barplot(x='Electric Range', y='Model', hue='Make', data=top_range_models, palette="cool")
plt.title('Top 10 Models by Average Electric Range in Top Makes')
plt.xlabel('Average Electric Range (miles)')
plt.ylabel('Model')
plt.legend(title='Make', loc='center right')
```

The TESLA ROADSTER has the highest average electric range, with TESLA models (ROADSTER, MODEL S, MODEL X, and MODEL 3) dominating the top positions, indicating TESLA's vehicles generally have higher ranges. The CHEVROLET BOLT EV stands out among CHEVROLET models with a notably higher range than the VOLT and S-10 PICKUP. NISSAN's LEAF and CHEVROLET's SPARK are in the lower half, indicating more modest average ranges.
